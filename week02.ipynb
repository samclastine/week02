```python
df4['iso2'].fillna(df4['iso2'].mode()[0],inplace=True)
df4['e_prev_100k']=df4['e_prev_100k'].fillna(df4['e_prev_100k'].mean())
df4['e_prev_100k_lo']=df4['e_prev_100k_lo'].fillna(df4['e_prev_100k_lo'].mean())
df4['e_prev_100k_hi']=df4['e_prev_100k_hi'].fillna(df4['e_prev_100k_hi'].mean())
df4['e_prev_num']=df4['e_prev_num'].fillna(df4['e_prev_num'].mean())
df4['e_prev_num_lo']=df4['e_prev_num_lo'].fillna(df4['e_prev_num_lo'].mean())
df4['e_prev_num_hi']=df4['e_prev_num_hi'].fillna(df4['e_prev_num_hi'].mean())
df4['e_mort_exc_tbhiv_100k']=df4['e_mort_exc_tbhiv_100k'].fillna(df4['e_mort_exc_tbhiv_100k'].mean())
df4['e_mort_exc_tbhiv_100k_lo']=df4['e_mort_exc_tbhiv_100k_lo'].fillna(df4['e_mort_exc_tbhiv_100k_lo'].mean())
df4['e_mort_exc_tbhiv_100k_hi']=df4['e_mort_exc_tbhiv_100k_hi'].fillna(df4['e_mort_exc_tbhiv_100k_hi'].mean())
df4['e_mort_exc_tbhiv_num']=df4['e_mort_exc_tbhiv_num'].fillna(df4['e_mort_exc_tbhiv_num'].mean())
df4['e_mort_exc_tbhiv_num_lo']=df4['e_mort_exc_tbhiv_num_lo'].fillna(df4['e_mort_exc_tbhiv_num_lo'].mean())
df4['e_mort_exc_tbhiv_num_hi']=df4['e_mort_exc_tbhiv_num_hi'].fillna(df4['e_mort_exc_tbhiv_num_hi'].mean())
```
